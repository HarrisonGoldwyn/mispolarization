```python
def sig_prime(omega, t): 
    return np.real(np.exp(1j*(omega)*t*1e-15 - dom.g(t, test_script_d)))
```

```python
omegas = np.linspace(0, .3/hbar, 10)
time_points = 10000
t_bound = 250
for w in omegas:
    plt.plot(
        np.linspace(0, t_bound, time_points), 
        sig_prime(w, np.linspace(0, t_bound, time_points)),
        label=f'w = {(hbar*w):.2f})'
        )
#     plt.xlim([0,.1])
        
plt.legend()
```
